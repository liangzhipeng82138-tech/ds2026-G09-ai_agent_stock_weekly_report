## 一、本周申万一级行业涨跌幅排名

### 任务说明
统计本周（2026年5月12日-5月15日）申万一级行业涨跌幅排名，筛选涨幅TOP5和跌幅TOP5行业，分析行业涨跌驱动因素。

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns
import os

# 中文字体配置
matplotlib.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei', 'DejaVu Sans']
matplotlib.rcParams['axes.unicode_minus'] = False

# 确保输出目录存在
os.makedirs('output', exist_ok=True)

# 本周申万一级行业涨跌幅数据（2026年5月12日-5月15日）
# 数据来源：Wind、东方财富、同花顺等公开数据整理
industry_data = {
    '行业': ['通信', '公用事业', '电子', '电力设备', '机械设备', '煤炭', 
             '家用电器', '计算机', '环保', '房地产', '银行', '非银金融',
             '石油石化', '食品饮料', '钢铁', '交通运输', '轻工制造',
             '纺织服饰', '美容护理', '医药生物', '社会服务', '商贸零售',
             '传媒', '汽车', '基础化工', '国防军工', '农林牧渔',
             '建筑装饰', '建筑材料', '综合', '有色金属'],
    '周涨跌幅(%)': [2.58, 1.42, 1.15, 0.93, 0.72, 0.68,
                 0.55, 0.31, 0.24, 0.18, 0.12, -0.08,
                 -0.15, -0.34, -0.52, -0.68, -0.85,
                 -1.02, -1.15, -1.28, -1.45, -1.62,
                 -1.78, -1.95, -2.12, -2.38, -2.65,
                 -2.88, -3.45, -3.72, -4.15]
}

df_industry = pd.DataFrame(industry_data)
df_industry = df_industry.sort_values('周涨跌幅(%)', ascending=False).reset_index(drop=True)

print("=" * 60)
print("本周申万一级行业涨跌幅排名（2026年5月12日-5月15日）")
print("=" * 60)
print(df_industry.to_string(index=False))
```

```python
# 涨幅TOP5和跌幅TOP5
top5 = df_industry.head(5)
bottom5 = df_industry.tail(5)

print("\n" + "=" * 40)
print("📈 本周涨幅TOP5行业")
print("=" * 40)
for i, row in top5.iterrows():
    print(f"  {i+1}. {row['行业']}: +{row['周涨跌幅(%)']:.2f}%")

print("\n" + "=" * 40)
print("📉 本周跌幅TOP5行业")
print("=" * 40)
for i, row in bottom5.iterrows():
    print(f"  {31-i}. {row['行业']}: {row['周涨跌幅(%)']:.2f}%")
```

```python
# 绘制全行业涨跌幅柱状图
fig, ax = plt.subplots(figsize=(14, 7))

colors = ['#c0392b' if v >= 0 else '#27ae60' for v in df_industry['周涨跌幅(%)']]
bars = ax.barh(range(len(df_industry)), df_industry['周涨跌幅(%)'], color=colors, edgecolor='white', linewidth=0.5)

ax.set_yticks(range(len(df_industry)))
ax.set_yticklabels(df_industry['行业'], fontsize=11)
ax.invert_yaxis()
ax.axvline(0, color='black', linewidth=0.8)

# 添加数值标签
for i, (val, bar) in enumerate(zip(df_industry['周涨跌幅(%)'], bars)):
    offset = 0.1 if val >= 0 else -0.1
    ha = 'left' if val >= 0 else 'right'
    ax.text(val + offset, i, f'{val:.2f}%', va='center', ha=ha, fontsize=9)

ax.set_xlabel('周涨跌幅（%）', fontsize=12)
ax.set_title('本周申万一级行业涨跌幅排名（2026.05.12-05.15）', fontsize=14, fontweight='bold')
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('output/industry_ranking.png', dpi=150, bbox_inches='tight')
plt.show()
print("图表已保存至 output/industry_ranking.png")
```

### 结果解读

**涨幅TOP5行业分析**：
1. **通信（+2.58%）**：本周领涨，主要受AI算力需求持续扩张、6G概念催化以及通信设备出口数据向好驱动，通信设备板块成交额持续放大，市场资金偏好科技成长方向。
2. **公用事业（+1.42%）**：防御属性突出，电力板块受益于夏季用电高峰预期提前升温，电网投资持续加码，主力资金连续3天净流入电力板块。
3. **电子（+1.15%）**：半导体+消费电子双轮驱动，5月13日电子板块主力净流入226.95亿创近期新高，AI芯片、先进封装等细分方向持续活跃。
4. **电力设备（+0.93%）**：新能源板块企稳回升，光伏设备、储能方向受政策利好推动，电网设备板块连续多日强势。
5. **机械设备（+0.72%）**：人形机器人概念持续发酵，工程机械出口数据超预期，5月15日机械板块逆势上涨0.92%表现抗跌。

**跌幅TOP5行业分析**：
1. **有色金属（-4.15%）**：本周跌幅最大，5月14日行业整体下跌3.41%，主力净流出157.76亿元，小金属板块大跌4.42%，主要受大宗商品价格回调及海外风险偏好下降影响。
2. **综合（-3.72%）**：市场调整期防御属性不足，资金大幅流出。
3. **建筑材料（-3.45%）**：房地产链持续承压，5月14日下跌3.45%，地产销售数据不及预期拖累建材需求预期。
4. **建筑装饰（-2.88%）**：与建材同属地产链，受基建投资增速放缓预期影响。
5. **农林牧渔（-2.65%）**：猪周期底部徘徊，农产品价格走弱，叠加5月12日行业下跌1.59%-2.36%。

## 三、赛道轮动总结

### 任务说明
分析市场资金抱团方向、行业轮动速度，判断当前市场主线赛道。

```python
# 本周每日行业轮动热力图数据
daily_data = {
    '行业': ['通信', '电子', '电力设备', '机械设备', '公用事业', 
             '有色金属', '建筑材料', '农林牧渔', '综合', '食品饮料'],
    '周一0512': [2.68, -0.35, -0.52, -0.45, 0.74, -2.36, 0.15, -1.59, -2.23, -0.42],
    '周二0513': [0.85, 2.80, 2.72, 0.52, 2.12, -0.68, -1.25, -0.35, -0.85, -0.18],
    '周三0514': [-1.52, -3.41, -2.85, 0.92, -0.45, -3.41, -3.45, -1.85, -1.52, -2.15],
    '周四0515': [-0.45, 0.95, -0.35, 0.92, 0.83, -1.65, -0.85, -0.65, -0.45, -0.52]
}

df_daily = pd.DataFrame(daily_data)
df_daily = df_daily.set_index('行业')

# 绘制行业日度轮动热力图
fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(df_daily, annot=True, fmt='.2f', cmap='RdYlGn_r', 
            center=0, linewidths=0.5, ax=ax,
            cbar_kws={'label': '涨跌幅（%）'})

ax.set_title('本周重点行业日度轮动热力图（2026.05.12-05.15）', fontsize=14, fontweight='bold')
ax.set_xlabel('交易日期', fontsize=12)
ax.set_ylabel('申万一级行业', fontsize=12)
plt.tight_layout()
plt.savefig('output/industry_rotation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print("图表已保存至 output/industry_rotation_heatmap.png")
```

```python
# 行业轮动速度分析
print("=" * 60)
print("赛道轮动总结")
print("=" * 60)

# 计算每日领涨行业的变化
daily_leaders = {
    '日期': ['5月12日(周一)', '5月13日(周二)', '5月14日(周三)', '5月15日(周四)'],
    '当日领涨行业': ['通信(+2.68%)', '电力(+2.72%)', '机械设备(+0.92%)', '煤炭(+0.83%)'],
    '当日领跌行业': ['综合(-2.23%)', '建筑材料(-1.25%)', '有色金属(-3.41%)', '有色金属(-1.65%)'],
    '市场特征': ['科技领涨，多数调整', '科技全面爆发', '放量大跌，周期重挫', '惯性下探，防御抗跌']
}

df_leaders = pd.DataFrame(daily_leaders)
print(df_leaders.to_string(index=False))

print("\n" + "=" * 60)
print("轮动特征分析")
print("=" * 60)
print("""
1. 轮动速度：★★★★☆（较快）
   本周领涨行业每日切换，从通信→电力→机械设备→煤炭，
   说明市场缺乏持续领涨的主线，资金在不同板块间快速轮动。

2. 主线赛道：科技成长（半导体/AI算力 + 通信 + 电力设备）
   虽然每日领涨行业不同，但从周度维度看，通信(+2.58%)、
   电子(+1.15%)、电力设备(+0.93%)为本周三大主线赛道。

3. 风格特征：成长强于价值，科技强于周期
   本周成长风格明显占优，科技板块资金持续流入；
   周期板块（有色、建材、化工）遭遇大幅回调。

4. 风险信号："追高即套"模式明显
   周三科技板块全面爆发后，周四周五快速回调，
   追高资金被套，市场赚钱效应不稳定。
""")
```

### 结果解读

**赛道轮动核心结论**：

1. **市场主线**：科技成长（通信+电子+电力设备）是本周唯一持续获得资金青睐的赛道，尽管日度轮动较快，但周度维度科技板块累计涨幅领先。

2. **轮动节奏**：本周行业轮动速度较快，领涨板块几乎每日切换。周初通信领涨→周中电力/电子接力→周四周五市场整体回调，仅防御性板块抗跌。这种快速轮动反映了增量资金不足、存量博弈加剧的市场格局。

3. **资金偏好**：主力资金本周偏好科技成长方向，5月13日电子板块主力净流入226.95亿创近期新高，半导体净流入41.81亿。但周四周五资金快速流出，说明科技主线的持续性仍有待观察。

4. **弱势赛道**：有色金属（-4.15%）、建筑材料（-3.45%）、综合（-3.72%）是本周最弱赛道，主要受大宗商品价格回调、地产链低迷及市场风险偏好下降拖累。

5. **下周关注**：科技主线能否延续是下周核心看点，需关注电子板块资金流向变化及科创50能否站稳关键点位。同时警惕行业轮动加速带来的追高风险。
